# Mobile Review Sentiment Analysis using Machine Learning

## Problem Statement

Mobile brands generate massive volumes of customer reviews across multiple platforms and regions, but relying solely on star ratings fails to capture true customer sentiment and the underlying reasons for satisfaction or dissatisfaction. This lack of actionable insight makes it difficult to identify key drivers of customer experience, understand market-specific perceptions, and make informed product and marketing decisions. There is a need for an automated, scalable approach to extract sentiment from review text, uncover drivers of dissatisfaction, and compare brand perception across markets.

## Business Questions

1. Can we automatically classify reviews into Positive, Neutral, and Negative?
2. Can sentiment be predicted accurately from review text alone?
3. What factors drive negative sentiment in mobile reviews?
4. Why do some low-rated phones still receive positive textual reviews?
5. How does sentiment vary by brand, price, country, and rating?
6. How does brand perception differ across regions (countries)?
7. Can we predict sentiment for new, unseen reviews in real time?

## Step 4 - Text Preprocessing

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words = set(stopwords.words('english'))
l = WordNetLemmatizer()
```

```python
def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z]', ' ',text)
    words = text.split()
    words = [l.lemmatize(w) for w in words if w not in stop_words]
    return ' '.join(words)

df['clean_review'] = df['review_text'].apply(clean_text)
```

```python
df['clean_review']
```

## Step 5 - Word Cloud Visualization

```python
positive_text = " ".join(df[df['sentiment']=='Positive']['clean_review'])
wordcloud = WordCloud(width=1000, height=900, background_color="white", colormap="Greens", max_words=200).generate(positive_text)
plt.figure(figsize=(8, 8))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

Insight: Positive sentiment strongly correlates with words like battery, camera, value and performance.

## Step 6 - Feature Engineering

```python
tfidf = TfidfVectorizer(max_features=5000)
X = tfidf.fit_transform(df['clean_review'])
y = df['sentiment']
```

```python
tfidf
```

## Step 7 - Train-Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

## Step 8 - Model Building

```python
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

## Step 9 - Model Evaluation

```python
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
plt.figure(figsize = (8,6))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap = 'coolwarm')
plt.title("Confusion Matrix")
plt.show()
```

#### Performance Results
- Accuracy: ~87%
- Strong precision for Positive & Negative
- Neutral sentiment is hardest (expected)

 Insight: Sentiment can be predicted accurately from text alone.

## Step 10 - Predict Sentiment for New Reviews

```python
def predict_sentiment(review):
    review = clean_text(review)
    vector = tfidf.transform([review])
    return model.predict(vector)[0]

print("Sentiment : ", predict_sentiment("Camera quality is excellent but battery drains fast"))
```

## Business Insights 

1. Customer ratings do not always reflect true sentiment, making textual review analysis essential for accurately understanding user opinions.

2. Battery life and device performance are the primary drivers of negative sentiment in mobile phone reviews.

3. Premium-priced phones receive more critical sentiment, as customers have higher expectations from high-end devices.

4. Brand perception varies significantly across countries, highlighting the impact of regional preferences and market conditions.

5. Positive textual reviews can still appear alongside low star ratings, indicating that ratings alone can be misleading.

6. Natural Language Processing models can reliably automate sentiment classification with high accuracy.

7. The trained sentiment model enables real-time prediction of customer sentiment for new and unseen reviews.
